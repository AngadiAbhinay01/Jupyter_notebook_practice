```python
class Node:
    def __init__(self,value):
        self.value = value
        self.ref = None
class sLinkedList:
    def __init__(self):
        self.head = None
    
    def traverse(self):
        if self.head is None:
            print("SLL is not exists")
        else:
            node = self.head
            while node is not None:
                print(node.value,"--->",end = "")
                node = node.ref
                
    #insertion
    def insertAtBegin(self,value):
        newNode = Node(value)
        newNode.ref = self.head
        self.head = newNode
                
    def insertAtEnd(self,value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
        else:
            node = self.head
            while node.ref is not None:
                node = node.ref
            node.ref = newNode
            
    def insertAfterNode(self,value,x):
        node = self.head
        while node is not None:
            if x == node.value:
                break
            node = node.ref
        if node is None:
            print("node is not present")
        else:
            newNode = Node(value)
            newNode.ref = node.ref
            node.ref = newNode 
            
    def insertBeforeNode(self,value,x):
        if self.head is None:
            print('Linked list is empty')
            return
        if self.head.value == x:
            newNode = Node(value)
            newNode.ref = self.head
            self.head = newNode
            return
        node = self.head
        while node.ref is not None:
            if node.ref.data == x:
                break
            node = node.ref
        if node.ref is None:
            print("node is not present")
        else:
            newNode = Node(value)
            newNode.ref = node.ref
            node.ref = newNode 
            
    def insertEmpty(self,value):
        if self.head is None:
            newNode = Node(value)
            self.head = newNode
        else:
            print("linked list is not present")
            
            
            
            
            
    def deleteBegin(self):
        if self.head is None:
            print("no node is present for deletion")
        else:
            self.head = self.head.ref
    
    def deleteEnd(self):
        if self.head is None:
            print("no node is present for deletion")
        else:
            node = self.head
            while node.ref.ref is not None:
                node = node.ref
            node.ref = None
    def deleteByValue(self,x):
        if self.head is None:
            print("can't delete node in linked list")
            return
        if x == self.head.value:
            self.head = self.head.ref
            return
        node = self.head
        while node.ref is not None:
            if x == node.ref.value:
                break
            node = node.ref
            
        if node.ref is None:
            print("node is not present")
        else:
            node.ref = node.ref.ref
    
            
            
singlyLinkedList = sLinkedList()
singlyLinkedList.insertAtBegin(10)
singlyLinkedList.insertAtEnd(100)
singlyLinkedList.insertAtBegin(8)

#singlyLinkedList.insertAfterNode(200,100)
#singlyLinkedList.insertBeforeNode(200,10)
#singlyLinkedList.insertEmpty(10)
#singlyLinkedList.insertEmpty(20)
#singlyLinkedList.insertEmpty(30)

#singlyLinkedList.deleteBegin()

#singlyLinkedList.deleteEnd()

singlyLinkedList.deleteByValue(100)



singlyLinkedList.traverse()


```
